### 5) Exploratory Data Analysis

```python
# Convert column 'Date' to Datetime :
dg['Date']= pd.to_datetime(dg['Date'])
ds['Date']= pd.to_datetime(ds['Date'])
```

```python
# Convert Date column as index :
dg= ds.set_index('Date')
ds = ds.set_index('Date')
```

```python
# Check the Updated DataFrames :
print('<------ GOLD ------>')
print(dg,'\n')
print('<------ SILVER ------>')
print(ds)
```

### 30-Day Moving Average Of Silver Prices

```python
# Calculate the 30-day moving average
window = 30
ds['30-Day MA'] = ds['Close/Last'].rolling(window=window).mean()

# Create a pivot table for the heatmap
heatmap_data = ds.pivot_table(index=ds.index.year, columns=ds.index.month, values='30-Day MA')

# Set up the heatmap figure
plt.figure(figsize=(15, 10))
plt.title('30-Day Moving Average of Silver Prices Over Time')

# Define your custom color palette with shades of silver and greys
custom_palette = sns.color_palette(["#C0C0C0", "#A0A0A0", "#808080", "#606060"], as_cmap=True)

# Create the heatmap using the custom color palette
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap=custom_palette, cbar_kws={'label': 'Moving Average'})
plt.xlabel('Month')
plt.ylabel('Year')

plt.show()
```

### To Compare The Long-Term Trends With The Short-Term Fluctuations Of Silver

```python
# Define short-term and long-term periods for moving averages
short_term_window = 30  
long_term_window = 180  

# Calculate short-term and long-term moving averages for silver
ds['Short-Term MA'] = ds['Close/Last'].rolling(window=short_term_window).mean()
ds['Long-Term MA'] = ds['Close/Last'].rolling(window=long_term_window).mean()

# Plot the silver price data along with short-term and long-term moving averages
plt.figure(figsize=(12, 6))

# Silver
plt.plot(ds.index, ds['Close/Last'], label='Silver Price', color='silver')
plt.plot(ds.index, ds['Short-Term MA'], label=f'Short-Term MA ({short_term_window} days)', color='blue')
plt.plot(ds.index, ds['Long-Term MA'], label=f'Long-Term MA ({long_term_window} days)', color='black')
plt.ylabel('Silver Price')
plt.title('Silver Price with Moving Averages')
plt.legend()

plt.tight_layout()
plt.show()
```

### A Heatmap Showing Average Monthly Gold Prices Across Different Years

```python
# Extract the year and month from the Date column
dg['Year'] = dg.index.year
dg['Month'] = dg.index.month

# Calculate the average gold price for each month and year
average_gold_prices = dg.groupby(['Year', 'Month'])['Close/Last'].mean().unstack()

# Define a custom color palette with five different shades of golden or related colors
custom_palette = sns.color_palette(['#FFD700', '#FFB90F', '#FFA500', '#FF8C00', '#FF4500'])

# Create the heatmap with the custom color palette
plt.figure(figsize=(12, 6))
sns.heatmap(average_gold_prices, cmap=custom_palette, annot=True, fmt=".2f", linewidths=.5)
plt.title('Average Gold Prices by Month and Year')
plt.xlabel('Month')
plt.ylabel('Year')

plt.tight_layout()
plt.show()
```

### Average Daily Volatility For Gold

```python
# Calculating the volatility (difference between high and low prices) for Gold :
dg['Volatility'] = dg['High'] - dg['Low']
```

```python
# Calculating the average daily volatility for Gold :
average_gold_volatility = dg['Volatility'].mean()
```

```python
# Creating a DataFrame to hold the average volatility data
average_volatility_data = pd.DataFrame({
    'Metal': ['Gold'],
    'Average Daily Volatility': [average_gold_volatility]})
average_volatility_data
```

```python
# Set the rolling window size:
rolling_window = 30

# Calculate rolling standard deviation of volatility:
dg['Rolling Volatility Std'] = dg['Volatility'].rolling(window=rolling_window).std()

# Create a figure with a single subplot for gold:
plt.figure(figsize=(12, 6))

# Plot daily volatility and rolling standard deviation for Gold:
plt.plot(dg.index, dg['Volatility'], label='Daily Volatility (Gold)', color='gold', alpha=0.5)
plt.plot(dg.index, dg['Rolling Volatility Std'], label=f'{rolling_window}-Day Rolling Std Dev (Gold)', color='red')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.title('Volatility and Rolling Standard Deviation in Gold Prices')
plt.legend()

plt.tight_layout()
plt.show()
```
